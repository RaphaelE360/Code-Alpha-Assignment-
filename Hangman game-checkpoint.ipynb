```python
import random

#list of predefined words
words = ['apple','stable','pride','grape','happy']

#randomly choosing a word
secret_word = random.choice(words)

#create hidden version of word using underscore
guessed_word = ["_ "] * len(secret_word)

#tracked guessed letters and wrong attempts
guessed_letters = []
wrong_guesses = 0
max_wrong = 6

print("Welcome to Hangman!")
print("Guess the word one letter at a time")

#game loop
while wrong_guesses < max_wrong and "_ " in guessed_word:
    print("\nWord:","".join(guessed_word))
    print("Guessed letters:", guessed_letters)
    print("Your guesses remaining:", max_wrong - wrong_guesses)
    
    guess = input("Enter a letter:").lower()

    #check if input is valid
    if len(guess) != 1 or not guess.isalpha():
        print("KIndly enter only one letter.")
        continue
    
    #check if already guessed
    if guess in guessed_letters:
        print("\n you already guessed that letter.")
        continue
        
    guessed_letters.append(guess)
        
    #check if guess is correct
    if guess in secret_word:
        print("correct")
        
        #reveal matching letters
        for i in range(len(secret_word)):
            if secret_word[i] == guess:
                guessed_word[i] = guess
                
    else:
        print("Wrong guess!")
        wrong_guesses += 1


#end of game
if "_ " not in guessed_word:
    print("\nCongratulations! You guessed the word:", secret_word)
else:
    print("\nGame Over! The word was:", secret_word)
```
